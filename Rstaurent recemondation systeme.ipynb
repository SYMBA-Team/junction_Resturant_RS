```python
import pandas as pd
```

```python
data= pd.read_csv('Classeur5.csv')
```

```python
data
```

```python
rest = data['food_name']
```

```python
r=rest.tolist()
```

```python
r
```

```python
chunks = [r[x:x+3] for x in range(0, len(r), 3)]
```

```python
chunks
```

```python
import mlxtend 
from mlxtend.preprocessing import TransactionEncoder 
te=TransactionEncoder() 
te_ary=te.fit(chunks).transform(chunks)    
df=pd.DataFrame(te_ary, columns=['chees_burger','double_chees_burger','beef_burger','chiken_sandiwch','tacos','hot_dog','coca_cola','0.5L_bottel_water','1L_bottel_water','Orange_juce'])  #Creating a new DataFrame from our Numpy array df
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.1)
```

```python
frequent_itemsets=apriori(df, min_support=0.1, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3) 
```
